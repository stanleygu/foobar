```python
# coding: utf-8


# In[2]:

import simclient


# In[ ]:

import urllib
import json
import tellurium as te
import numpy as n
get_ipython().magic(u'matplotlib inline')



link = "https://raw.githubusercontent.com/stanleygu/biomodels_db/master/curated/BIOMD0000000012.xml"
f = urllib.urlopen(link)
sbml = f.read()
```

```python
import notebooktools.notebooktools as nb

rr = te.loadSBMLModel(sbml)
rr.model.getGlobalParameterIds()
nb.makeParameterSliders(rr)
```

```python
job = {
    'sbml': {
        'string': sbml,
        'name': 'MySim'
    },
    'time': {
        'start': 0,
        'end': 100,
        'steps': 100
    },
    'params': {
        'n': 1
    }
}
for item, i in enumerate(range(4)):
    job['sbml']['name'] += str(i)
    job['params']['n'] = i
    result = simclient.addSimJob(job)
    output = json.loads(result)
    te.plotArray(n.array(output))
```
