```python
def makeParameterSliders(r,
                         paramIds=None,
                         minFactor=0,
                         maxFactor=2,
                         sliderStepFactor=10):
    import matplotlib.pyplot as plt
    %matplotlib inline
    from IPython.html.widgets import interact, interactive, fixed
    from IPython.html import widgets
    from IPython.display import clear_output, display, HTML
    import sys

    if paramIds is None:
        paramIds = r.model.getGlobalParameterIds()
    paramMap = {}

    def runSim(start=0, stop=100, steps=100, **paramMap):
        r.reset()
        for k,v in paramMap.items():
            try:
                key = k.encode('ascii', 'ignore')
                r.model[key] = v
            except:
                # error in setting model variable
                e = sys.exc_info()
                print e

        try: 
            s = r.simulate(start, stop, steps)
            te.plotArray(s)
        except:
            # error in simulation
            e = sys.exc_info()
            print e


    for i, id in enumerate(paramIds):
        val = r.model[id]
        try:
            r.model[id] = val
            paramMap[id] = widgets.FloatSliderWidget(
                        min=minFactor*val, max=maxFactor*val, step=val/sliderStepFactor, value=val)
        except:
            e = sys.exc_info()
            print e

    i = interact(runSim,
             start=widgets.FloatTextWidget(min=0, value=0),
             stop=widgets.FloatTextWidget(min=0, value=100),
             steps=widgets.IntTextWidget(min=0, value=100),
             **paramMap
             )
```

```python
import tellurium as te
r = te.loadSBMLModel("Glycolysis.xml")

makeParameterSliders(r, paramIds=['J1_k1'])
```
