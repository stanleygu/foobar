```python
import tellurium as te
```

```python
model1 = '''
model example2
  S1 -> S2 + S3; k1*S1 #Mass-action kinetics
  S1 = 10  #The initial concentration of S1
  S2 = 0   #The initial concentration of S2
  S3 = 3   #The initial concentration of S3
  k1 = 0.1 #The value of the kinetic parameter from J0.
end
'''

model2 = '''
model example2
  J0: S1 -> S2 + S3; k1*S1 #Mass-action kinetics
  S1 = 10  #The initial concentration of S1
  S2 = 0   #The initial concentration of S2
  S3 = 3   #The initial concentration of S3
  k1 = 0.1 #The value of the kinetic parameter from J0.
  k1' = J0
end
'''
```

```python
te.loadAntimonyModel(model1)
sbml1 = te.antimonyTosbml(model1)
print sbml1
r2 = te.loadAntimonyModel(model2)
sbml2 = te.antimonyTosbml(model2)
print sbml2
```

```python
%matplotlib inline
r2.reset()
result = r2.simulate()
p = te.plotWithLegend(r2, result)
p.axes
```
